```python
#set up dependencies

import pandas as pd

pymoli_data = "Resources/purchase_data.csv"

pymoli_df = pd.read_csv(pymoli_data)

pymoli_df.head()
```

```python
pymoli_df.count()
```

```python
# calculating total number of players

player= len(pymoli_df["SN"].value_counts())
player_count = pd.DataFrame([player], columns = ["Total Players"])
player_count
```

```python
# calculating purchasing analysis figures

unique = len(pymoli_df["Item Name"].unique())
average_price = pymoli_df["Price"].mean()
no_of_purchases = len(pymoli_df["Item Name"])
total_revenue = pymoli_df["Price"].sum()
purchasing_analysis = pd.DataFrame({ "Number of Unique Items": [unique],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [no_of_purchases],
                                    "Total Revenue": [total_revenue]
                                   })

# formatting average price and total revenue columns

purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].map("${:.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:.2f}".format)
purchasing_analysis
```

```python
# calculating gender demographics

gender = pymoli_df[["SN", "Gender"]]
gender = gender.drop_duplicates()

initial_count = gender["Gender"].value_counts()

total_count = [initial_count[0], initial_count[1], initial_count[2]]

percentage = [round((initial_count[0]/player)*100, 2),
             round((initial_count[1]/player)*100, 2),
             round((initial_count[2]/player)*100, 2)]

gender_demographics = pd.DataFrame({"Gender": ["Male", "Female", "Other/Non-disclosed"],
                                   "Percentage of Players": percentage,
                                    "Total Count": total_count
                                   })

# formatting percentage column

gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:,.2f}%".format)
gender_demographics
```

```python
# calculating purchasing analysis (gender)

gender_group = pymoli_df.groupby(["Gender"])

purchase_count = gender_group["SN"].count()
avg_purchase_price = gender_group["Price"].mean()
purchase_value = gender_group["Price"].sum()

drop_duplicates = pymoli_df.drop_duplicates(subset="SN", keep="first")
grouped_duplicates = drop_duplicates.groupby(["Gender"])

average_purchase_pp = (gender_group["Price"].sum()/grouped_duplicates["SN"].count())

gender_purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                        "Average Purchase Price": avg_purchase_price,
                                        "Total Purchase Value": purchase_value,
                                        "Average Purchase Per Person": average_purchase_pp
                                        })
# formatting columns

gender_purchase_analysis["Average Purchase Price"] = gender_purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_purchase_analysis["Total Purchase Value"]= gender_purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
gender_purchase_analysis["Average Purchase Per Person"] = gender_purchase_analysis["Average Purchase Per Person"].map("${:.2f}".format)

gender_purchase_analysis
```

```python
# set bins for age demographics
dup_free_df = pymoli_df.drop_duplicates("SN")

dup_free_df["Age"].max() # used to determine max age

bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

dup_free_df["Age Demographics"] = pd.cut(dup_free_df["Age"], bins, labels=group_names, include_lowest=True)
age_df = dup_free_df.groupby(["Age Demographics"])
initial_count = dup_free_df["Age Demographics"].value_counts()
total_count = [initial_count[6], initial_count[5], initial_count[2], initial_count[0], initial_count[1], initial_count[3], initial_count[4], initial_count[7]]
age_percentage = [round((initial_count[6]/player)*100, 2),
                  round((initial_count[5]/player)*100, 2),
                  round((initial_count[2]/player)*100, 2),
                  round((initial_count[0]/player)*100, 2),
                  round((initial_count[1]/player)*100, 2),
                  round((initial_count[3]/player)*100, 2),
                  round((initial_count[4]/player)*100, 2),
                  round((initial_count[7]/player)*100, 2)]

age_demographics = pd.DataFrame({"Age": ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"],
                                 "Total Count": total_count,
                                 "Percentage of Players": age_percentage
                                })

age_demographics = age_demographics.set_index("Age")
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:,.2f}%".format)
age_demographics
```

```python
# calculating age purchase analysis

age_avg_price = age_df["Price"].mean()
age_total_value = age_df["Price"].sum()
avg_purchase_pp = age_total_value/initial_count


age_purchase_analysis = pd.DataFrame({"Purchase Count": initial_count,
                                     "Average Purchase Price": age_avg_price,
                                     "Total Purchase Value": age_total_value,
                                     "Average Purchase Per Person": avg_purchase_pp
                                     })

# format the dataframe

age_purchase_analysis["Average Purchase Price"] = age_purchase_analysis["Average Purchase Price"].map("${:,.2f}".format)
age_purchase_analysis["Total Purchase Value"] = age_purchase_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_purchase_analysis["Average Purchase Per Person"] = age_purchase_analysis["Average Purchase Per Person"].map("${:,.2f}".format)

age_purchase_analysis
```

```python
# calculating top spenders

sn_group = pymoli_df.groupby(["SN"])

user_purchase_count = sn_group["Price"].count()
avg_user_purchase = sn_group["Price"].mean()
total_user_purchase = sn_group["Price"].sum()

top_spender = pd.DataFrame({"Purchase Count": user_purchase_count,
                        "Average Purchase Price": avg_user_purchase,
                        "Total Purchase Value": total_user_purchase
                        })

# formatting data frame

sorted_top_spender = top_spender.sort_values("Total Purchase Value", ascending=False)

sorted_top_spender["Average Purchase Price"] = sorted_top_spender["Average Purchase Price"].map("${:,.2f}".format)
sorted_top_spender["Total Purchase Value"] = sorted_top_spender["Total Purchase Value"].map("${:,.2f}".format)

sorted_top_spender.head()
```

```python
# calculating most popular items

popular_item = pymoli_df.set_index(["Item ID", "Item Name"])
popular_item = popular_item.groupby(level = ["Item ID", "Item Name"])

popular_purchase_count = popular_item["Price"].count()
popular_item_price = popular_item["Price"].mean()
popular_total_value = popular_item["Price"].sum()

popular_items_df = pd.DataFrame({"Purchase Count": popular_purchase_count,
                                "Average Item Price": popular_item_price,
                                "Total Purchase Value": popular_total_value
                                })

# formatting data frame

sorted_popular_items = popular_items_df.sort_values("Purchase Count", ascending=False)

sorted_popular_items["Average Item Price"] = sorted_popular_items["Average Item Price"].map("${:,.2f}".format)
sorted_popular_items["Total Purchase Value"] = sorted_popular_items["Total Purchase Value"].map("${:,.2f}".format)

sorted_popular_items.head()
```
